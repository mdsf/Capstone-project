```python
import pandas as pd # import library to read data into dataframe
pd.set_option('display.max_columns', None)
import numpy as np # import numpy library
```

```python
spambase = pd.read_csv("https://www.openml.org/data/get_csv/44/dataset_44_spambase.arff")

print("Data read into dataframe!") # takes about 30 seconds
```

# Dividing Data into Spam and Non Spam E-mail Categories

```python
df = pd.DataFrame(spambase)
#print(f_result['class'])
not_spam = df.loc[df['class'] == 0]
spam = df.loc[df['class'] == 1]
```

Some prelimary info about the data set

Class Column denotes whether the e-mail was considered spam (1) or not (0)

48 continuous real [0,100] attributes of type

word_freq_WORD = percentage of words in the e-mail that match WORD

# DataFrame containing Spam e-mail data

```python
#spam.head()
count_spam_values = spam.iloc[:, 1:].sum(axis=0)
count_spam_values
```

```python
used_word_money = spam.groupby("word_freq_money").mean()
used_word_money
```

# DataFrame containing Non-Spam e-mail data

```python
count_not_spam_values = not_spam.iloc[:, 1:].sum(axis=0)
count_not_spam_values
```

```python
used_word_money_ns = not_spam.groupby("word_freq_money").mean()
used_word_money_ns
```
