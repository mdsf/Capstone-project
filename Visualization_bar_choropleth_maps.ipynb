```python
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data')

print('Dataset downloaded and read into a pandas dataframe!')
```

```python
df_survey.rename(columns={'Unnamed: 0':''}, inplace=True)
df_survey.set_index('',inplace=True)
df_survey.head()
```

```python
df_survey.sort_values(by='Very interested', ascending=False, inplace=True)
df_survey = (df_survey/2233)
df_survey
```

```python
ax = df_survey.plot(kind='bar', width=0.8, figsize=(20, 8), color = ('#5cb85c', '#5bc0de', '#d9534f'),fontsize=14)
ax.set_title('Percentage of Respondents Interest in Data Science Areas',fontsize=16)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.get_yaxis().set_ticks([])
ax.patches
#for p in ax.patches:
    #ax.annotate("{:.02%}".format(p.get_height()),
               # xy=(p.get_x(), p.get_height()))
```
